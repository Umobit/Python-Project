```python
# Convert the booking_time from the dataset to datetime
bok['booking_time'] = pd.to_datetime(bok['booking_time'])

#Bring out the days of the week from the conversion above
bok['Days_of_week'] = bok['booking_time'].dt.day_name()

# Count the number of bookings for each day of the week
boki_by_day = bok['Days_of_week'].value_counts()

#Question 3
print (boki_by_day, '\n\nThursday has the highest number of bookings')
```

```python
# Plot a pie chart of the Bookings Distribution by day of the week
plt.figure(figsize=(10, 10))
boki_by_day.plot.pie(autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Bookings Distribution for all Days of the Week')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
customer_booking_counts = bok ['customer_id'].value_counts()
multiple_bookings_customers = customer_booking_counts[customer_booking_counts > 1].index

# Filter the dataframe to include only these customers
filtered_bok = bok[bok['customer_id'].isin(multiple_bookings_customers)]

# Create a new column for city
filtered_bok['City'] = filtered_bok['from_city'] + ' to ' + filtered_bok['to_city']

# Group by 'Customer ID' and 'city' and count the number of bookings
route_counts = filtered_bok.groupby(['customer_id', 'City']).size().reset_index(name='booking_count')


# Identify the most booked route for each customer
most_booked_route = route_counts.loc[route_counts.groupby('customer_id')['booking_count'].idxmax()]

# Display the result
print(most_booked_route)
```

```python
customers_multiple_bookings = bok.groupby('customer_id').filter(lambda x: len(x) > 1)
most_booked_route = customers_multiple_bookings.groupby(['from_city', 'to_city']).size().idxmax()

print(f'The most booked route for customers with more than one booking are: {most_booked_route}')
```

```python
# Question 6, Which are the top 3 departure cities from where customers book mostly in advance, provided that there have been at least 5 departures from that city?

#bok['booking_time'] = pd.to_datetime(bok['booking_time'])
bok['days_to_departure'] = pd.to_datetime(bok['days_to_departure'])

#bok['days_in_advance'] = (bok['departure_date'] - bok['booking_date']).dt.days

top_departure_cities = bok.groupby('from_city').filter(lambda x: len(x) >= 5)
top_departure_cities = top_departure_cities.groupby('from_city')['days_to_departure'].mean().nlargest(3)

print(top_departure_cities)
```

```python
numerical_cols = bok.select_dtypes(include=[np.number])

corr_matrix = numerical_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap=plt.cm.Paired_r, fmt='.1f', annot_kws={"size": 15})
plt.title('Heatmap Correlation of Numerical Columns')
plt.show()

corr_matrix_unstacked = corr_matrix.unstack()
sorted_corr = corr_matrix_unstacked.sort_values(ascending=False)


sorted_corr = sorted_corr[sorted_corr < 1]

print("Pair of columns with the highest correlation:")
print(sorted_corr.idxmax(), sorted_corr.max())
```

```python
# 8) For each service, which is the most used device type for making bookings on the platform?

service_device_counts = bok.groupby(['service_name', 'device_type_used']).size().reset_index(name='booking_count')

# For each service, find the device type with the highest booking count
most_used_device_per_service = service_device_counts.loc[service_device_counts.groupby('service_name')['booking_count'].idxmax()]

# Display the result
print("Most used device type for each service:")
print(most_used_device_per_service)
```

```python
#9) Plot the trends at a quarterly frequency for the number of bookings by each of the device types, that is, plot a time series for each year and quarter showing the number of bookings performed by each device type.

bok['booking_time'] = pd.to_datetime(bok['booking_time'], errors='coerce')


bok['quarter'] = bok['booking_time'].dt.to_period('Q')


quarterly_trends = bok.groupby(['quarter', 'device_type_used']).size().unstack().fillna(0)


quarterly_trends.plot(figsize=(12, 8))
plt.title('Quarterly Trends for Number of Bookings by Device Type')
plt.xlabel('Quarter')
plt.ylabel('Number of Bookings')
plt.legend(title='Device Type')
plt.show()
```

```python
total_searches = 100
total_bookings = 10
total_customers = 12


oBSR = total_bookings / total_searches


data = {
    'Date': pd.date_range(start='2024-01-01', periods=total_searches + total_bookings, freq='D')
}
activity_df = pd.DataFrame(data)

# Calculate oBSR for each day
activity_df['oBSR'] = total_bookings / (activity_df.index + 1)


monthly_avg_oBSR = activity_df.groupby(activity_df['Date'].dt.month)['oBSR'].mean()
daily_avg_oBSR = activity_df.groupby(activity_df['Date'].dt.dayofweek)['oBSR'].mean()

# Plot the time series of oBSR
plt.figure(figsize=(12, 6))
plt.plot(activity_df['Date'], activity_df['oBSR'], marker='o', linestyle='-')
plt.title('Time Series of oBSR')
plt.xlabel('Date')
plt.ylabel('oBSR')
plt.grid(True)
plt.show()

# Print average oBSR for each month and day of the week
print("Average oBSR for each month:")
print(monthly_avg_oBSR)
print("\nAverage oBSR for each day of the week:")
print(daily_avg_oBSR)
```
